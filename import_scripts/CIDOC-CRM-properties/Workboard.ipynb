```python
import pandas as pd
from wikidataintegrator import wdi_core, wdi_login
from getpass import getpass
import numpy as np
```

```python
df = pd.read_excel (r'Export_SAF_LUX_10_dead can dance.xlsx')
df
```

```python
wikibase = "https://cidocoption1a.wiki.opencura.com/"
api = "https://cidocoption1a.wiki.opencura.com/w/api.php"
sparql = "https://cidocoption1a.wiki.opencura.com/query/sparql"
entityUri = wikibase.replace("https:", "http:")+"entity/"

WBUSER = "Andrawaag"
WBPASS = getpass()
login = wdi_login.WDLogin(WBUSER, WBPASS, mediawiki_api_url=api)
```

```python
for index, row in df.iterrows():
    statements = []
    statements.append(wdi_core.WDItemID(value="Q30", prop_nr="P2"))
    statements.append(wdi_core.WDMonolingualText(value=row["Heading:Name"], prop_nr="P8"))
    label_name = row["Heading:Name"]
    # Numeration
    if not pd.isnull(row["Heading:Numeration"]):
        statements.append(wdi_core.WDItemID(value=numeral_qids[row["Heading:Numeration"]], prop_nr="P10"))
        label_name += " "+row["Heading:Numeration"]
    # Title (Caveat: Only english as lang tag now)               
    if not pd.isnull(row["Heading:Title"]):
        statements.append(wdi_core.WDMonolingualText(value=row["Heading:Title"], prop_nr="P11")) 
        label_name += " "+row["Heading:Title"]
    # Alternative names
    if not pd.isnull(row["Alternative / other form(s) of name:Name"]):
        aliases =  row["Alternative / other form(s) of name:Name"].split(";")
        for altName in aliases:
            statements.append(wdi_core.WDMonolingualText(value=altName, prop_nr="P8"))
    person_item = wdi_core.WDItemEngine(new_item=True, data=statements, mediawiki_api_url=api, sparql_endpoint_url=sparql)
    
    person_item.set_label(label_name, lang="en")
    person_item.set_label(label_name, lang="fr")
    person_item.set_label(label_name, lang="de")
    if len(aliases)>0:
        person_item.set_aliases(aliases, lang="en")
        person_item.set_aliases(aliases, lang="fr")
        person_item.set_aliases(aliases, lang="de")
    print(person_item.write(login))

        
```

```python
# Numerations
query = """
    PREFIX wd: <http://cidocoption1a.wiki.opencura.com/entity/>
    PREFIX wdt: <http://cidocoption1a.wiki.opencura.com/prop/direct/>

    SELECT ?item ?itemLabel WHERE {
       ?item wdt:P2 wd:Q338 .
      SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
    }

"""
numeral_qids = dict()
df_numerals = wdi_core.WDItemEngine.execute_sparql_query(query=query, endpoint=sparql, as_dataframe=True)
for index, row in df_numerals.iterrows():
    numeral_qids[row["itemLabel"]] = row["item"].replace("http://cidocoption1a.wiki.opencura.com/entity/", "")
    print(row["item"].replace("http://cidocoption1a.wiki.opencura.com/entity/", ""), row["itemLabel"])
```
